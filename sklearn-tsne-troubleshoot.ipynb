```python
import numpy as np
from sklearn.manifold import TSNE

temp = '5 3 15 '
```

```python
for i in range(20):
    for j in range(5):
        string = "; {} {} {} ".format(str(i),str(j),str(j*i))
        temp += string
        
    
```

```python
temp = np.matrix(temp)
```

```python
tsn = TSNE(n_components=3)
tsn.fit_transform(temp)
```
